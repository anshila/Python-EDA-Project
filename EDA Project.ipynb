####  <a id="question1">1. How many are there in each Team and the percentage splitting with respect to the total employees.</a>


```python
# Number of employees in each team
team_count = df['Team'].value_counts()
team_count
```

```python
# Percentage splitting of the total employees
percentage_splitting = round((team_count / len(df)) * 100,2)
percentage_splitting
```

```python
df['percentage_splitting'] = df['Team'].map(percentage_splitting)
df
```

```python
# Create a new DataFrame with the results
team_percentage_splitting = pd.DataFrame({'Team': team_count.index, 'Number of Employees': team_count.values, 'Percentage Splitting': percentage_splitting.values})
team_percentage_splitting
```

The data reveals the distribution of employees across various teams, showcasing team sizes and their respective percentages of the total workforce. The 'New Orleans Pelicans' and 'Memphis Grizzlies' lead with 19 and 18 employees, respectively, while other teams maintain a relatively consistent workforce percentage, ranging from 3.06% to 4.15%. 

```python
# Plotting the bar chart 
plt.figure(figsize=(12,6))
plt.xticks(rotation="vertical")
plt.title('Percentage of Employees in Each Team')
plt.xlabel('Team')
plt.ylabel('Percentage')
sns.countplot(x='Team',data=df,hue='Team',palette='plasma')
plt.show()
```

#### <a id="question3">3. Find from which age group most of the employees belong to.</a>


```python
df['Age']
```

```python
plt.figure(figsize=(10,6))
sns.countplot(x='Age',data=df,hue='Age',palette='rocket')
plt.title('Age Distribution of Employees')
plt.xlabel('Age')
plt.ylabel('No: of Employees')
plt.show()
```

<b>The age distribution chart provides a visual representation of the workforce within the organization, showcasing the count of employees in each age group.</b><br><br>
Notably, the highest concentration of employees falls within the 22 to 29 age range, indicating a substantial presence of early to mid-career professionals. <br>

```python
#Age group 
bins = [15,20,25,30,35,40,45]
labels = ['15-20','21-25','26-30','31-35','36-40','41-45']
df['Age Group'] = pd.cut(df['Age'], bins=bins, labels=labels, right=False)
df['Age Group'].value_counts()
```

```python
# Plot age group distribution
plt.figure(figsize=(10, 6))
sns.countplot(x='Age Group', data=df, hue='Age Group',palette='rocket')
plt.title('Age Group Distribution of Employees')
plt.xlabel('Age Group')
plt.ylabel('Number of Employees')
plt.show()
```

The countplot function is employed to create a bar chart, with the x-axis representing the age groups and the y-axis indicating the number of employees. This visualization provides a clear and engaging representation of the age distribution within the workforce, facilitating easy interpretation and analysis of the data.

The highest count is observed in the 26-30 age group, indicating a significant presence of employees in their mid-career stage. Following closely, the 21-25 and 31-35 age groups emerge as the second and third highest, respectively. The 36-40 and 41-45 age groups represent the next tiers in the workforce. This breakdown helps us understand the diverse mix of experiences and career stages within the organization.


#### <a id="question4">4. Find out under which team and position, spending in terms of salary is high.</a>


```python
x = df['Salary'].mean()
df['Salary'].fillna(x,inplace = True)
df1=df[['Team','Position','Salary']]
df1
```

```python
# Group by Team and Position, then sum the salaries
team_position_salary = df1.groupby(['Team', 'Position'])['Salary'].sum()
team_position_salary
```

```python
# Find the team and position with the highest spending
highest_spending_position = team_position_salary.idxmax()
highest_salary = team_position_salary.loc[highest_spending_position]
team, position = highest_spending_position
team, position,highest_salary
```

```python
# Plotting the bar plot
plt.figure(figsize=(12, 6))
sns.barplot(x='Team', y='Salary', hue='Position', data=df1)
plt.title('Highest Spending Salary by Team and Position')
plt.xlabel('Team')
plt.ylabel('Total Salary')
plt.xticks(rotation=45)
plt.show()
```

The bar plot visualizes the highest spending salary by team and position. Each bar represents the total salary for a specific team, differentiated by positions. The highest spending position is highlighted for each team, providing a clear comparison of salary distribution.
eam.

* **X-axis (Horizontal Axis):** The x-axis represents the "Team." Each bar on the plot corresponds to a specific team in the dataset. <br>
* **Y-axis (Vertical Axis):** The y-axis represents the "Total Salary." The height of each bar corresponds to the total salary for a particular team and position combination. <br>
* **Hue (Color):** The hue (color) represents the "Position." Different positions within each team are distinguished by different colors. This additional dimension helps in understanding how the total salary is distributed among various positions within each team.

#### <a id="question5">5. Find if there is any correlation between age and salary , represent it visually</a>


```python
df2=df[['Age','Salary']]
df2
```

```python
#Calculate the correlation matrix
correlation= df2.corr()
correlation
```

```python
sns.heatmap(correlation, annot=True, cmap="crest")
plt.title("Correlation Matrix between Age & Salary of Employee Dataset")
plt.show()
```

The correlation matrix for "Age" and "Salary" indicates a positive correlation. This value suggests a weak positive linear relationship between age and salary. While the correlation is statistically significant, the strength of the association is relatively low. Therefore, as age increases, there is a tendency for salary to also increase, but the effect is not particularly strong.
